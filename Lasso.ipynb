```python
import pandas as pd
import numpy as np
df=pd.read_csv("Details_Cars.csv")
df
```

```python
df=df.replace("?",np.nan)
df
```

```python
col_obj=df.select_dtypes(include=["object"])
col_obj
```

```python
df["price"]=pd.to_numeric(df["price"],errors="coerce")
df["peak-rpm"]=pd.to_numeric(df["peak-rpm"],errors="coerce")
df["horsepower"]=pd.to_numeric(df["horsepower"],errors="coerce")
df["stroke"]=pd.to_numeric(df["stroke"],errors="coerce")
df["bore"]=pd.to_numeric(df["bore"],errors="coerce")
```

```python
col_obj=df.select_dtypes(include=["object"])
col_obj
```

```python
df=df.drop("normalized-losses",axis=1)
df
```

```python
df.dtypes
```

```python
cylin={"two":2,"three":3,"four":4,"five":5,"six":6,"eight":8,"twelve":12}
df["num-of-cylinders"].replace(cylin,inplace=True)
df
```

```python
df["num-of-cylinders"].head()
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
df.dtypes
```

```python
df=df.dropna()
df
```

```python
corr=df.corr()
corr
```

```python
X=df.drop("price",axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
print(model.score(X_test,y_test))
```

```python
predictors=X_train.columns
coef=pd.Series(model.coef_,predictors).sort_values()
```

```python
predictors
```

```python
coef
```

```python
import matplotlib.pyplot as plt
coef.plot(kind="bar")
```

```python
from sklearn.linear_model import Lasso
lasso_model=Lasso(alpha=1,normalize=True)
lasso_model.fit(X_train,y_train)
```

```python
print(lasso_model.score(X_test,y_test))
```

```python
predictors=X_train.columns
lasso_coef=pd.Series(lasso_model.coef_,predictors).sort_values()
```

```python
lasso_coef.plot(kind="bar")
```
